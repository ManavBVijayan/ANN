```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
```

```python
import keras
```

```python
import pandas as pd
```

```python
X=pd.DataFrame(x)
y=pd.DataFrame(y)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,Dropout,LeakyReLU,PReLU
```

```python
classifier=Sequential()
```

```python
X.shape
```

```python
classifier.add(Dense(units=4,activation='relu',input_dim=1,kernel_initializer='he_uniform'))
```

```python
classifier.add(Dense(units=4,activation='relu',kernel_initializer='he_uniform'))
```

```python
classifier.add(Dense(units=1,activation='linear',kernel_initializer='glorot_uniform'))
```

```python
classifier.compile(optimizer="AdaMax",loss="mean_squared_error",metrics=["mae"])
```

```python
classifier.fit(X,y,epochs=500)
```

```python
import numpy as np
new_data_point = np.array([6]).reshape(-1, 1)
```

```python
classifier.predict(new_data_point)
```
